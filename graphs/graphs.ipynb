```python
%matplotlib inline

import matplotlib
import numpy as np
import matplotlib.pyplot as plt
import pylab
```

```python
# Overall Course Rating
x_list = [9, 2, 0, 0, 0]
label_list = ["Excellent", "Very Good", "Good", "Fair", "Poor"]
 
plt.axis("equal")
plt.pie(
        x_list,
        autopct="%1.1f%%"
        )
#plt.title("Overall Course Rating")
plt.legend(label_list, loc='upper left')
plt.savefig("1overall.png")
plt.show()
```

```python
# University Part
x_list = [3, 4, 2, 1]
label_list = ["Department of Economics", "Harris School of Public Policy", "Booth School of Business", "Other"]
 
plt.axis("equal")
plt.pie(
        x_list,
        autopct="%1.1f%%"
        )
#plt.title("Which part of the University are you from?")
plt.legend(label_list, loc='center left', bbox_to_anchor=(0.9, 0.5))
plt.savefig("2part.png", bbox_inches='tight')
plt.show()
```

```python
# Usefulness
x_list = [0, 0, 1, 2, 7]
label_list = ["Not useful at all: 1", "2", "3", "4", "Very useful: 5"]

N = len(x_list)
x = range(N)
x_middle_ticks = np.arange(0.5,5.5)
width = 1/1.5
plt.bar(x, x_list, color="blue")

#plt.title("Relative to your other classes, how useful was this course?", y=1.08, fontsize=14)
#plt.text(label_list, loc='center left', bbox_to_anchor=(1, 0.5))
#plt.text(5.5, 6, r'1: Not useful at all', fontsize=13)
#plt.text(5.5, 5, r'2', fontsize=13)
#plt.text(5.5, 4, r'3', fontsize=13)
#plt.text(5.5, 3, r'4', fontsize=13)
#plt.text(5.5, 2, r'5: Very useful', fontsize=13)

plt.axis([0,5,0,8])


plt.xticks(x_middle_ticks,  label_list, horizontalalignment='center',
        verticalalignment='top', y=-0.01,
        rotation='vertical')


plt.savefig("3useful.png", bbox_inches='tight')

plt.show()
```

```python
# Well organized?
x_list = [9,1,0,0,0]
pos = np.arange(0, 10, 2)+.5
label_list = ["Excellent", "Very Good", "Good", "Fair", "Poor"]

#plt.title("to be well organized? (How well did Philipp meet his goal...)", y=1.08, fontsize=14)
#plt.text(label_list, loc='center left', bbox_to_anchor=(1, 0.5))
plt.barh(pos,x_list, align='center')
plt.yticks(pos, label_list)

# plt.axis([0,10, 0, 10])
plt.savefig("4organized.png", bbox_inches='tight')
plt.show()
```
